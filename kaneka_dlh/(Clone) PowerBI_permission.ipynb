```python
def get_account():
 print("ユーザー情報を取得しています…")
 user = spark.sql("SELECT current_user()").collect()[0][0]
 print(f"true:ユーザー{user}")
 return(user)
```

```python
def permission_check():
    print("権限内容を取得しています…")
    df = spark.sql("""
        SELECT *
        FROM dev_first_usecase.gold.user_table
        WHERE user_id = current_user()""")
    print("権限内容を取得しました")
```

```python
def view_delete():
    if spark.catalog.tableExists("dev_first_usecase.gold.GOLD_VIEW"):
        print("ビューをreplaceします")
        spark.sql(f"""
            CREATE OR REPLACE VIEW dev_first_usecase.gold.GOLD_VIEW AS
            SELECT g.*
            FROM dev_first_usecase.gold.GOLD AS g
            JOIN  (
                SELECT DISTINCT cost_center, master
               FROM dev_first_usecase.gold.user_table
               WHERE user_id = current_user()
            ) AS u
            ON (u.master = TRUE OR g.cost_center = u.cost_center);
        """)
        print("true:ビューをreplaceしました")
```

```python
def view_create():
    if not spark.catalog.tableExists("dev_first_usecase.gold.GOLD_VIEW"):
        print("ビューが存在しません、新規作成を行います")
        spark.sql(f"""
            CREATE OR REPLACE VIEW dev_first_usecase.gold.GOLD_VIEW AS
            SELECT g.*
            FROM dev_first_usecase.gold.GOLD AS g
            JOIN  (
                SELECT DISTINCT cost_center, master
               FROM dev_first_usecase.gold.user_table
               WHERE user_id = current_user()
            ) AS u
            ON (u.master = TRUE OR g.cost_center = u.cost_center);
        """)
        print("true:ビュー作成が完了しました")
```

```python
try:
    get_account()
    permission_check()
    view_delete()
    view_create()
    print("TRUE:PowerBIからビューを指定して読み込んでください")
except Exception as e:
    raise Exception(f"エラーが発生しました：{e}")
```
